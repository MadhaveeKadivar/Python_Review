```python
import pandas as pd
import re

df = pd.read_csv("test.csv")

mo_no = df.iloc[:,-1]
def func(str):
    """ 
        Description: 
            This function is printing different wrong format mobile number in correct format
        Parameter:
            It takes one string as argument
        Return:
            returns correct mobile number string 
    """
    if(re.match("^\d{10}$",str)):
        str = "+91"+str
        return str
    if(re.match("^91[\s]?\d{10}$",str)):        
        str = str.replace(str[:-10],"+91")
        return str
    if(re.match("^0[\s]?\d{10}$",str)):
        str = "+91" + str[-10:]
        return str
    if(re.match("^[+]91[\s]\d{10}$" ,str)):
        str = str.replace(str[3],"")
        return str
    if(re.match("^\d{11,}$",str)):
        str = "+91" + str[0:10]
        return str
    return str

def full_name(x , y):
    """ 
        Description: 
            This function is giving full name based on first name and last name
        Parameter:
            It takes two string as argument
        Return:
            returns full name string
    """
    return x + " " + y

df['full_name'] = df.apply(lambda x : full_name(x['fname'],x['lname']),axis = 1)

df['mnumber'] = df['mnumber'].apply(func)

df
```
